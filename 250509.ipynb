```python
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.plot(months, rainfall,'.--')
for i in range(len(months)):
  plt.annotate(rainfall[i],(months[i],rainfall[i]), xytext=(0,5), textcoords='offset points')
plt.title('Monthly average rainfall')
plt.xlabel('Month')
plt.ylabel('rainfall')
plt.show()
```

```python
#y축 범위 설정 (0부터 시작하여 )
plt.ylim(0,max(rainfall)*1.1)

# 데이터 포인터에 값 표시
for i, value in enumerate(rainfall):
  plt.text(i, value + 10, str(value), ha='center', va='bottom')

#레이아웃 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```
