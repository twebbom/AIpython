## **08장 연습문제**

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
temperatures = [5, 7, 12, 18, 22]

plt.plot(months, temperatures)
plt.title('Monthly Temperature Change')
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.show()
```

```python
height = [160, 165, 170, 175, 180]
weight = [55, 60, 68, 72, 78]

plt.scatter(height, weight)
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
# 예시 데이터: 과일별 판매량
fruits = ['Apple', 'Banana', 'Orange']
sales = [100, 150, 80]

plt.bar(fruits, sales) # x축 카테고리, y축 값
plt.title('Fruit Sales')
plt.xlabel('Fruit')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
# 예시 데이터: 학생들의 시험 점수
scores = [65, 72, 75, 78, 80, 82, 85, 88, 90, 91, 93, 95, 98, 100]

plt.hist(scores, bins=5) # 데이터, 구간(bin) 개수 설정
plt.title('Exam Score Distribution')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
import matplotlib.pyplot as plt

year = [2006, 2009, 2012, 2015, 2018]
kor = [547, 546, 554, 524, 526]
usa = [474, 487, 481, 470, 478]
plt.plot(year, kor)
plt.plot(year, usa)
plt.show()
```

```python
import matplotlib.pyplot as plt

year = [2006, 2009, 2012, 2015, 2018]
kor = [547, 546, 554, 524, 526]
usa = [474, 487, 481, 470, 478]
jpn = [523, 529, 536, 532, 527]

# 그래프가 가로와 세로 크기를 각각 5인치와 3인치 크기 설정
plt.figure(figsize=(5,3))
# year 데이터를 x축으로, kor 데이터를 y축으로 하여 빨간색 점선으로 선 그래프를 그리고 범례에 'Korea'라고 표시
plt.plot(year, kor, 'r:', label='Korea')
plt.plot(year, usa, 'b:', label='USA')
plt.plot(year, jpn, 'g-', label='Japan')
plt.ylim(400, 600)  # y축 범위 설정
# 그래프에 범례를 추가
plt.legend()
# 그래프 제목을 'Mathematics performance'로 설정합니다.
plt.title('Mathmatics performance')
plt.xlabel('year')
plt.ylabel('score')
plt.show()
```

### 강의노트 연습 문제

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.plot(months, rainfall)
plt.title('Monthly average rainfall')
plt.xlabel('Month')
plt.ylabel('rainfall')
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.figure(figsize=(12,6))
plt.plot(months, rainfall, 'r:.')
plt.title('Monthly average rainfall')
plt.xlabel('Month')
plt.ylabel('rainfall')
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.scatter(months, rainfall)
for i in range(len(months)):
  plt.annotate(rainfall[i],(months[i],rainfall[i]), xytext=(5,-10), textcoords='offset points')
plt.title('Monthly average rainfall')
plt.xlabel('Month')
plt.ylabel('rainfall')
plt.show()
```
