# 실습문제 딕셔너리

### 문제 1: 딕셔너리 생성 및 요소 접근

* 학생 정보를 담는 딕셔너리 `student`를 생성합니다. 키로는 `'name'`,`'age'`,`'major'`를 사용하고, 각각에 적절한 값을 할당합니다.
* 딕셔너리의 각 키를 사용하여 해당 값을 접근합니다.
* 접근한 각 키와 값을 "키:값" 형태로 `print()` 함수를 사용하여 출력합니다.

```python
student = {'name':'김민수', 'age':20, 'major':'컴퓨터공학'}
print(f'학생 정보 딕셔너리: {student}')

name = student['name']
print(f'이름: {name}')

age = student['age']
print(f'나이: {age}')

major = student['major']
print(f'전공: {major}')
```

### 문제 2: 딕셔너리 요소 추가 및 삭제

* 문제 1에서 생성한 `student` 딕셔너리에 새로운 키 `'grade'`와 값 `4.0`을 추가합니다. 딕셔너리에 새로운 키-값 쌍을 할당하는 방식으로 추가합니다.
* 추가된 후의 딕셔너리를 출력합니다.
* 딕셔너리에서 `'major'` 키-값 쌍을 삭제합니다. `del` 키워드를 사용합니다.
* 삭제된 후의 딕셔너리를 출력합니다.

```python
student = {'name':'김민수', 'age':20, 'major':'컴퓨터공학'}
print(f'원래 학생 정보 딕셔너리: {student}')

student['grade'] = 4.0 # 새로운 키 'grade'와 값 추가
print(f'요소 추가 후 딕셔너리: {student}')

del student['major'] # 키 'major'와 값 삭제
print(f'요소 삭제 후 딕셔너리: {student}')
```

### 문제 3: 딕셔너리 키와 값 순회하며 출력

* 과일과 그 가격을 담는 딕셔너리 `fruit_prices`를 생성합니다.
* `for` 루프와 딕셔너리의 `items()` 메서드를 사용하여 모든 키-값 쌍을 순회합니다.
* 각 키와 값을 "과일: 가격" 형태로 `print()` 함수를 사용하여 출력합니다.
* `for` 루프와 딕셔너리의 `keys()` 메서드를 사용하여 모든 키를 순회하고 출력합니다.
* `for` 루프와 딕셔너리의 `values()` 메서드를 사용하여 모든 값을 순회하고 출력합니다.

```python
fruit_prices = {'사과': 1500, '바나나':800, '딸기':3000, '오렌지': 1200}

print('과일 가격 정보: ')
for fruit, price in fruit_prices.items():
  print(f'{fruit}: {price}원')

print('\n과일 목록:')
for fruit in fruit_prices.keys():
  print(fruit)

print('\n가격 목록:')
for price in fruit_prices.values():
  print(f'{price}원')
```

### 문제 4: 딕셔너리 요소 값 변경

* 문제 3에서 생성한 `fruit_prices` 딕셔너리에서 '사과'의 가격을 새로운 값으로 변경합니다. 딕셔너리의 특정 키에 새로운 값을 할당하는 방식으로 수정합니다.
* 변경 전과 변경 후의 딕셔너리를 각각 출력하여 결과를 확인합니다.

```python
fruit_prices = {'사과': 1500, '바나나':800, '딸기':3000, '오렌지': 1200}
print(f'원래 과일 가격 딕셔너리: {fruit_prices}')

fruit_prices['사과'] = 1800 # '사과'의 가격을 1800으로 변경
print(f'가격 변경 후 과일 가격 딕셔너리: {fruit_prices}')
```

### 문제 5: 두 개의 딕셔너리 합치기

* 첫 번째 딕셔너리 `dict1`을 생성하고 몇 개의 키-값 쌍을 초기 값으로 할당합니다.
* 두 번째 딕셔너리 `dict2`를 생성하고 몇 개의 키-값 쌍을 초기 값으로 할당합니다. (일부 키가 `dict1`과 겹치도록 설정해봅니다.)
* `update()` 메서드를 사용하여 `dict1`에 `dict2`의 모든 키-값 쌍을 추가합니다. 이때, 키가 겹치는 경우 `dict2`의 값으로 업데이트 됩니다.
* 합쳐진 딕셔너리를 출력합니다.

```python
dict1 = {'a':1, 'b':2, 'c':3}
dict2 = {'b':4, 'd':5}
print(f'첫 번째 딕셔너리: {dict1}')
print(f'두 번째 딕셔너리: {dict2}')

dict1.update(dict2) #dict의 내용을 dict2에 업데이트 (겹치는 키는 dict2의 값으로 덮어쓰기)
print(f'두 딕셔너리 합치기 (update() 결과): {dict1}')
```

# 코딩 실습2

BMI(Body Mass Index)는 키와 몸무게로 계산된 체지방 지수이다. 흥미롭게도 이 BMI 기준을 만든 건 의사도 과학자도 아닌 미국의 보험업체인 메트로폴리탄이다. 대상자의 키(m)가 H, 몸무게가 W(kg.)이라고 할 때 BMI는 다음과 같이 계산된다.

BMI  =  W  /  H ^ 2

예를 들어, 몸무게 55kg에 키가 1.68m인 사람의 BMI는  = 19.48696 이다. 여러분은 어떤 사람의 키와 몸무게를 입력 받아 이 사람의 BMI 값을 출력해야 한다. (소수 셋째 자리 반올림, round() 함수 사용)

<참고>
- input() 함수의 입력값은 문자열이므로, float() 함수를 이용하여 실수형으로 변환할 수 있습니다.  예를 들어, tall = float(input())   입니다.
- round() 함수는 특정 자리에서 반올림을 해 줍니다. 예를 들어, 소수 셋째 자리에서 반올림을 하는 경우 round(bmi, 2) 로 사용할 수 있습니다.


입력
첫째 줄에 키 (cm), 둘째 줄에 몸무게 (kg) 가 주어진다.

출력
BMI 를 출력한다.

```python
h = float(input())/100
w = float(input())

bmi = w/h**2
print(round(bmi, 2))
```
